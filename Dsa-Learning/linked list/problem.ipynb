## Remove every k-th node of the linked list

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
def delete_k(head, k):
    if  head is None or k <= 0:
        return head
    curr = head
    prev = None
    count = 0
    while curr is not None:
        count += 1
        if count % k == 0:
            if prev is not None:
                prev.next = curr.next
            else:
                head = curr.next
        else:
            prev = curr
        curr = curr.next
    return head
def print_list(head):
    curr = head
    while curr is not None:
        print(curr.data, end=" -> ")
        curr = curr.next
    print("None")
if __name__ == "__main__":
  
    # Create a hard-coded linked list:
    # 1 -> 2 -> 3 -> 4 -> 5 -> 6
    head = Node(1)
    head.next = Node(2)
    head.next.next = Node(3)
    head.next.next.next = Node(4)
    head.next.next.next.next = Node(5)
    head.next.next.next.next.next = Node(6)
    k = 2
    head = delete_k(head, k)
    print_list(head)
```

## Middle node in a linked list 

```python
class Node:
    def __init__(self, x):
        self.data = x
        self.next = None

def getLength(head):

    length = 0
    while head:
        length += 1
        head = head.next

    return length

def getMiddle(head):

    length = getLength(head)

    
    midIndex = length // 2
    while midIndex:
        head = head.next
        midIndex -= 1

    return head.data


if __name__ == "__main__":

    head = Node(10)
    head.next = Node(20)
    head.next.next = Node(30)
    head.next.next.next = Node(40)
    head.next.next.next.next = Node(50)
    head.next.next.next.next.next = Node(60)

    print(getMiddle(head))
```

## swaping first node and last node

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
def swap_first_last(head):
    if head is None or head.next is None:
        return head
    prev = None
    curr = head
    while curr.next is not None:
        prev = curr
        curr = curr.next
    curr.next = head.next
    prev.next = head
    head.next = None
    head, curr = curr, head
    return head
def print_list(head):
    curr = head
    while curr is not None:
        print(curr.data, end=" -> ")
        curr = curr.next
    print("None")
if __name__ == "__main__":
    head = Node(1)
    head.next = Node(32)
    head.next.next = Node(322)
    head.next.next.next = Node(24)
    head.next.next.next.next = Node(521)
    print("Original list:")
    print_list(head)
    head = swap_first_last(head)
    print("List after swapping first and last nodes:")
    print_list(head)
```

## Remove Duplicates from Sorted Linked list

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

    def remove_duplicates(head):
        curr  = head
        while curr and curr.next:
            if curr.data == curr.next.data:
                curr.next = curr.next.next
            else:
                curr = curr.next
        return head
def print_list(head):
    curr = head
    while curr is not None:
        print(curr.data, end=" -> ")
        curr = curr.next
    print("None")
if __name__ == "__main__":
    head = Node(10)
    head.next = Node(10)
    head.next.next = Node(20)
    head.next.next.next = Node(30)
    head.next.next.next.next = Node(30)
    head.next.next.next.next.next = Node(30)
    head.next.next.next.next.next.next= Node(40)

    print("Original list:")
    print_list(head)

    head = Node.remove_duplicates(head)

    print("List after removing duplicates:")
    print_list(head)
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def removeDuplicatesUnsorted(head):
    # if list is empty or single node
    if head is None:
        return head

    seen = set()        # store unique values
    current = head
    prev = None

    while current:
        if current.data in seen:
            # duplicate found — skip it
            prev.next = current.next
        else:
            # not duplicate — add to set
            seen.add(current.data)
            prev = current
        current = current.next

    return head


def printList(head):
    while head:
        print(head.data, end=" -> ")
        head = head.next
    print("None")


if __name__ == "__main__":
  
    head = Node(10)
    head.next = Node(20)
    head.next.next = Node(10)
    head.next.next.next = Node(30)
    head.next.next.next.next = Node(20)
    head.next.next.next.next.next = Node(40)

    print("Original Unsorted Linked List:")
    printList(head)

    head = removeDuplicatesUnsorted(head)

    print("\nLinked List After Removing Duplicates:")
    printList(head)
```

## Rotate the sorted linked list

```python
class Node:
    def __init__(self, newData):
        self.data = newData
        self.next = None

def rotate(head, k):
    if k == 0 or head is None:
        return head

    # rotate the list by k nodes
    for _ in range(k):
        curr = head
        while curr.next is not None:
            curr = curr.next
        
        # move the first node to the last
        curr.next = head
        curr = curr.next
        head = head.next
        curr.next = None
    
    return head

def printList(node):
    while node is not None:
        print(node.data, end = "")
        if node.next != None:
            print(" -> ", end = "")
        node = node.next
    print()

if __name__ == "__main__":

    head = Node(10)
    head.next = Node(20)
    head.next.next = Node(30)
    head.next.next.next = Node(40)

    head = rotate(head, 6)
    printList(head)
```
